```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def quadratic_function(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
file = open("output.txt", "r")
c = float(file.readline())
b = float(file.readline())
a = float(file.readline())

x = np.linspace(-2, 4, 100)
y = quadratic_function(x, a, b, c)
plt.plot(x, y)

x_points = [float(i) for i in file.readline().split()]
y_points = [float(i) for i in file.readline().split()]
plt.plot(x_points, y_points, 'o', color='red')

file.close()

plt.title(f'Đồ thị hàm số y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
