```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
```

```python
file_result = open('result_SIM.txt', 'r')
x_true = np.array([float(x) for x in file_result.readline().split()])
file_result.close()

file_output = open('output_SIM.txt', 'r')
lines = file_output.readlines()
x_approx = []
for line in lines:
    x_approx.append([float(x) for x in line.strip().split()])
file_output.close()
x_approx = np.array(x_approx)

iterations = np.arange(len(x_approx))

len_x = x_true.shape[0]

plt.figure(figsize=(10, 8))

for i in range(len_x):
    plt.subplot(int(ceil(len_x/3)), 3, i+1)
    plt.plot(iterations, x_approx[:, i], marker='o', linestyle='-', label=f'Xấp xỉ x{i+1}')
    plt.axhline(y=x_true[i], color='r', linestyle='--', label=f'Thực x{i+1}')
    plt.xlabel('Số lần lặp')
    plt.ylabel(f'Giá trị x{i+1}')
    plt.title(f'Biểu đồ đường cho ẩn x{i+1}')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
