```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
```

```python
def quadratic_function(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
file = open("output_PI.txt", "r")
c = float(file.readline())
b = float(file.readline())
a = float(file.readline())

x_points = [float(i) for i in file.readline().split()]
y_points = [float(i) for i in file.readline().split()]

x = np.linspace(min(x_points)-1, max(x_points)+1, 100)
y = quadratic_function(x, a, b, c)

file.close()

a1, b1, c1 = np.polyfit(x_points, y_points, 2)

mymodel = np.poly1d([a1, b1, c1])

print(f'R2 score: {r2_score(y_points, mymodel(x_points))}')

plt.figure(figsize=(22, 10))
plt.subplot(1, 2, 1)
plt.scatter(x_points, y_points, color='red', label=f"Point")
plt.plot(x, y, color='blue', label=f"My method")
plt.title(f' th hm s y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(x_points, y_points, color='red', label=f"Point")
plt.plot(x, mymodel(x), color='green', label=f"Sklearn method")
plt.title(f' th hm s y = {a1}x^2 + {b1}x + {c1}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```
